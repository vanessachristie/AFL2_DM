```python
import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Data Mining\customer_analysis.csv', delimiter='\t')
df
```

```python
#data cleaning
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.loc[:, 'Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format="%d-%m-%Y")
df
```

```python
coba = (df.dtypes == 'object')
categorical = list(coba[coba].index)
print(f'Categorical Variable: {categorical}')
```

```python
#feature engineering
#tambahin kolom age
df['Age'] = 2023-df['Year_Birth']
df
```

```python
#tambahin kolom Spent
df['Spent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts']
+ df['MntSweetProducts'] + df['MntGoldProds']
df
```

```python
#tambahin kolom Living_With
df['Living_With'] = df['Marital_Status'].replace({'Married':'Partner','Together':'Partner',
                                                 'Absurd':'Alone','Widow':'Alone','YOLO':'Alone','Divorced':'Alone','Single':'Alone'})
df
```

```python
#tambahin kolom Children 
df['Children'] = df['Kidhome'] + df['Teenhome']
df
```

```python
#tambahin kolom Family_Size
df['Family_Size'] = df['Living_With'].replace({'Alone': 1, 'Partner': 2}) + df['Children']
df
```

```python
#tambahin kolom Is_Parent
df['Is_Parent'] = np.where(df.Children > 0,1,0)
df
```

```python
#Ubah kolom Education
df['Education'] = df['Education'].replace({'Basic':'Undergraduate','2nd Cycle':'Undergraduate','2n Cycle':'Undergraduate',
                                          'Graduation':'Graduate','Master':'Postgraduate','PhD':'Postgraduate'})
df
```

```python
drop_column = ["Marital_Status", "Kidhome", "Teenhome", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntGoldProds", "MntSweetProducts",  "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID",'AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','AcceptedCmp1','AcceptedCmp2','Complain','Response']
df = df.drop(drop_column, axis=1)
df
```

```python
#data preparation
from sklearn.preprocessing import LabelEncoder
df['Education'] = LabelEncoder().fit_transform(df['Education'])
df['Living_With'] = LabelEncoder().fit_transform(df['Living_With'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
reduced_data = pd.DataFrame(pca.fit_transform(scaled_df), columns=(["PCA1", "PCA2", "PCA3"]))
reduced_data.describe().T
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = reduced_data['PCA1']
y = reduced_data['PCA2']
z = reduced_data['PCA3']
# plot
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, c='blue',marker='x')
ax.set_title('Dimensionality Reduction',fontsize=20)
plt.show()
```

```python
#Clustering
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

elbow_d = KElbowVisualizer(KMeans(),k=10)
elbow_d.fit(reduced_data)
elbow_d.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=4)
yhat_ac = ac.fit_predict(reduced_data)
reduced_data['Cluster'] = yhat_ac
df['Cluster'] = yhat_ac
df['Count'] = df.groupby('Cluster')['Income'].transform('count').astype(int)
df.groupby('Cluster').mean()
```

```python
from mpl_toolkits.mplot3d import Axes3D

cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(18,12))
ax = plt.subplot(111, projection='3d',label='bla')
ax.scatter(x,y,z, s=40, c=reduced_data['Cluster'],marker='x',cmap=cmap)
ax.set_title('Cluster',fontsize=20)
plt.show()
```
